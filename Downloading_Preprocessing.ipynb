```python
data = pd.read_csv("Chest_xray_Corona_Metadata.csv", index_col=0)
```

```python
all_files = list(data['X_ray_image_name'])
```

```python
all_labels = list(data['Label'])
```

```python
len(all_labels)
```

```python
dataset_types = list(data['Dataset_type'])
```

```python
dataset_types[0].lower()
```

```python
plt.figure(figsize=(16, 8))
for i in range(16):
    rand = random.randint(0, len(all_files))
    img = plt.imread(os.path.join(data_dir, dataset_types[rand].lower(), all_files[rand]))
    plt.subplot(4, 4, i+1)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    plt.title(all_labels[rand])
plt.show()
```

```python
import seaborn as sns
```

```python
sx = sns.countplot(x=all_labels)
for cont in sx.containers:
    sx.bar_label(cont)
sx.set_title('Coronahack Chest XRay Dataset')
sx.set_ylabel('Number of images')
plt.show()
```
